```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df_py = pd.read_csv("mod10_input_DE_comp.txt.csv",index_col='Time')
df_py.head()
```

```python
df_lib=pd.read_table("../LIB_comp_trace.txt",sep=",",index_col='time')
df_lib.head()
```

```python
fig,axs = plt.subplots(3,1,figsize=(6,6))
ax = axs[0]
ax.plot(df_lib['O_2'],label='libbi')
ax.plot(df_py['O2'],label='python')

ax=axs[1]
ax.plot(df_lib['p_H'],label='libbi')
ax.plot(df_py['pH'],label='python')

ax=axs[2]
ax.plot(df_lib['HCO3'],label='libbi')
ax.plot(df_py['HCO3'],label='python')
ax.legend()
```
